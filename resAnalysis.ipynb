```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

flex_pos = pd.read_csv('./Const_adr_pos/Const_adr_2h/Const_adr_2h.txt')
Time = pd.date_range(start='15/10/2017',periods=4704,freq='H')
flex_pos.index = Time
```

```python
Qflex = flex_pos[['Qflex']].copy()
Qflex.loc[:,'Hour'] = flex_pos.index.hour

fig, ax = plt.subplots(1,1, figsize=[15,10])
Qflex['month'] = Qflex.index.month
sns.stripplot(x="Hour", y="Qflex", data=Qflex, ax=ax, hue='month')
Qflex.groupby('Hour').aggregate(np.median).plot(ax=ax)
```

```python
fig, ax = plt.subplots(1,1, figsize=[15,10])
sns.boxplot(x="Hour", y="Qflex", data=Qflex, ax=ax)
```

```python
flex_pos_sb = pd.read_csv('./Setback_adr_pos/Setback_adr_2h/Setback_adr_2h.txt',sep='\t')
flex_pos_sb.index = Time
Qflex_sb = flex_pos_sb[['Qflex']].copy()
Qflex_sb.loc[:,'Hour'] = flex_pos_sb.index.hour

fig, ax = plt.subplots(1,1, figsize=[15,10])
Qflex_sb['month'] = Qflex_sb.index.month
sns.swarmplot(x="Hour", y="Qflex", data=Qflex_sb, ax=ax, hue='month')
```
